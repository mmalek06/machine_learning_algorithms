```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)


X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1) # sampling from normal distribution makes the chart look more natural

plt.plot(X, y, 'b.')
plt.axis([0, 2, 0, 15])
plt.show()
```

The np.ones call is there because of: https://datascience.stackexchange.com/questions/27945/what-is-the-use-of-additional-column-of-1s-in-normal-equation

One could use either of the below approaches, though the second one is optimal.

```python
def linear_regression_normal_equation(X, y):
    X_b = np.c_[np.ones((X.shape[0], 1)), X]

    return np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
def linear_regression_svd(X, y):
    X_b = np.c_[np.ones((X.shape[0], 1)), X]

    return np.linalg.pinv(X_b).dot(y)
```

```python
result = linear_regression_svd(X, y)

print(result)
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_pred = X_new_b.dot(result)

plt.clf()
plt.plot(X_new, y_pred, 'r-')
plt.plot(X, y, 'b.')
plt.axis([0, 2, 0, 15])
plt.show()
```
